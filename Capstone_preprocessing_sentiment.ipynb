```python
import pandas as pd
import numpy as np
```

```python
df3 = pd.read_csv("C:\\Users\\Tamar\\Documents\\dfV5.csv")
test_companies = pd.read_csv("C:\\Users\\Tamar\\Documents\\MMA\\Capstone Project\\test_companies_V2.csv")
test_companies = test_companies[['conm','cik','ap','ch','icapt','invt','ivst','lct','re','rect','at','bankrupt']]
test_companies_manip = test_companies[['conm','ap','ch','icapt','invt','ivst','lct','re','rect','at','bankrupt']]
```

```python
#make bakrupted column
count = 0
for i in range(len(df3)):
    if df3.loc[i,'dlrsn'] == 2 or df3.loc[i,'dlrsn'] == 3:
        count+=1
        df3.loc[i,'bankrupt']= 1
    else:
        df3.loc[i,'bankrupt'] = 0
print(count)
```

```python
df3 = df3[['conm','ap', 'ch','icapt', 'invt', 'ivst', 'lct', 're', 'rect','at','bankrupt','curcd']]
```

```python
for i in range(len(df3)):
    if df3.loc[i,'curcd'] == 'CAD':
        df3.loc[i,'ap'] = df3.loc[i,'ap']/1.39
        df3.loc[i,'ch'] = df3.loc[i,'ch']/1.39
        df3.loc[i,'icapt'] = df3.loc[i,'invt']/1.39
        df3.loc[i,'lct'] = df3.loc[i,'lct']/1.39
        df3.loc[i,'re'] = df3.loc[i,'re']/1.39
        df3.loc[i,'rect'] = df3.loc[i,'rect']/1.39
        df3.loc[i,'at'] = df3.loc[i,'at']/1.39
        
df3 = df3.drop(columns='curcd')
```

```python
df3 = df3.append(test_companies_manip)
```

```python
df3.tail()
```

```python
numerical = df3[['ap', 'ch','icapt', 'invt', 'ivst', 'lct', 're', 'rect','at']]

#import imputing packages
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

#uses multiple regression to fill missing values in a round-robin way
imp = IterativeImputer(max_iter=10, verbose=0)
imp.fit(numerical)
imputed = imp.transform(numerical)
imputed_df = pd.DataFrame(imputed, columns=numerical.columns)
```

```python
df3['ap'] = imputed_df['ap']
df3['ch'] = imputed_df['ch']
df3['icapt'] = imputed_df['icapt']
df3['invt'] = imputed_df['invt']
df3['ivst'] = imputed_df['ivst']
df3['lct'] = imputed_df['lct']
df3['re'] = imputed_df['re']
df3['rect'] = imputed_df['rect']
df3['at'] = imputed_df['at']
```

```python
df3['acctsRecieve/at'] = df3['rect']/df3['at']
```

```python
df3 = df3.replace([np.inf, -np.inf], np.nan)
df3 = df3.dropna()
```
